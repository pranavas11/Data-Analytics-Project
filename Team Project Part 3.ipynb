# Data Cleaning

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

# making dictionaries to refer to states and their abbreviations
states_to_abbrevs = {}
abbrevs_to_states = {}
with open("states_abbrevs.txt") as file:
    lines = file.readlines()
    for line in lines:
        arr = line.split(',')
        abbrev, state = arr[0], arr[1].replace('\n','')
        states_to_abbrevs[state] = abbrev
        abbrevs_to_states[abbrev] = state
```

```python
url1 = "https://uselectionatlas.org/RESULTS/state.php?year=2020&fips="
url2 = "&f=1&off=0&elect=0"

LIMIT = 60
election_info = [] # list of counties, states, state abbreviations, and democrat voter %

for i in range(1,LIMIT):
    response = requests.get(url1 + str(i) + url2)
    if response.status_code != 200:
        raise exception("Aborted: Received status code",response.status_code)
    
    soup = BeautifulSoup(response.text, "html.parser")
    
    if soup.text.find("Error") != -1 or i == 2:
        continue

    # getting the state name
    state_tag = soup.find("title").text
    state = state_tag[state_tag.find("-")+2:]
    
    # getting the county names
    li_tags = soup.find_all("li")
    county_tags = li_tags[-2].contents[-4]
    county_tags = county_tags.find_all("option")
    curr_counties = [tag.text for tag in county_tags]
    
    # getting the democrat vote percentages
    state_map = soup.find("map")
    abbrev = state_map["id"][0:2]
    areas = state_map.find_all("area")
    for area in areas:
        county = area["alt"]
        alt_msg = area["onmouseover"]
        alt_msg = alt_msg[alt_msg.find("(D)"):]
        alt_msg = alt_msg[alt_msg.find("\'>")+2:]
        percent_dems = float(alt_msg[:alt_msg.find("%")])
        election_info.append([county, state, abbrev, np.round(0.01 * percent_dems,3)])

# adding alaska election info since it wasn't available on the same website
# source: https://rrhelections.com/index.php/2021/04/13/alaska-presidential-results-by-county-equivalent-1960-2020/9/
with open("alaska_election_info.txt") as ak:
    lines = ak.readlines()
    for line in lines:
        arr = line.split(',')
        arr[3] = float(arr[3].replace('\n',''))
        election_info.append(arr)

# adding election info for some independent cities in virginia since it wasn't available on the same website
# source: https://www.politico.com/2020-election/results/virginia/
with open("virginia_election_info.txt") as va:
    lines = va.readlines()
    for line in lines:
        arr = line.split(',')
        arr[3] = float(arr[3].replace('\n',''))
        election_info.append(arr)

election_info[:10]
```

```python
election_df = pd.DataFrame(election_info, columns=["County","State","Abbrev","Fraction of Democrat Voters in 2020"])

# small changes to ensure proper merging

# changing 'District of Columbia County' to 'District of Columbia'
dc_index = list(election_df[election_df.State == "District of Columbia"].index)[0]
election_df.iat[dc_index,0] = "District of Columbia"

# duplicate county in Georgia
election_df = election_df[election_df.index != 517].reset_index()
election_df = election_df.iloc[:,1:]

# duplicate county in Kentucky
election_df = election_df[election_df.index != 1000].reset_index()
election_df = election_df.iloc[:,1:]

# changing 'La Salle Parish' to 'LaSalle Parish'

la_index = list(election_df[election_df.County == "La Salle Parish"].index)[0]
election_df.iat[la_index,0] = "LaSalle Parish"

# changing 'Lac Qui Parle County' to 'Lac qui Parle County'
mn_index = list(election_df[election_df.County == "Lac Qui Parle County"].index)[0]
election_df.iat[mn_index,0] = "Lac qui Parle County"

# changing 'Carson City City' to 'Carson City'
nv_index = list(election_df[election_df.County == "Carson City City"].index)[0]
election_df.iat[nv_index,0] = "Carson City"

# changing 'Dona Ana County' to 'Doña Ana County'
nm_index = list(election_df[election_df.County == "Dona Ana County"].index)[0]
election_df.iat[nm_index,0] = "Doña Ana County"

election_df
```

```python
# Data from https://www.opportunityatlas.org/
csv_files = ["cty_med_hhinc2016_real","cty_hs_rP_gP_pall","cty_coll_rP_gP_pall","cty_jail_rP_gP_pall",
             "cty_teenbirth_rP_gF_pall","cty_popdensity2010","cty_poor_share2016"]
df1 = pd.read_csv(csv_files[0] + ".csv")

# getting and merging the data
for i in range(1,len(csv_files)):
    data = pd.read_csv(csv_files[i] + ".csv")
    df1 = pd.merge(df1, data, how="inner", on=["cty", "Name"])

# data cleanup
df1 = df1.sort_index()
df1 = df1.iloc[:,1:]
df1.columns = ["County","Median Household Income","HS Graduation Rate","College Graduation Rate","Incarceration Rate",
              "Teenage Birth Rate","Population Density","Poverty Rate"]
df1 = df1[df1.County.isna() == False]
df1 = df1[df1.County.str[-2:] != "PR"]
df1 = df1.reset_index()
df1 = df1.iloc[:,1:]

counties_abbrevs = list(df1.County.str.split(','))
counties = []
states = []
abbrevs = []
for i in range(len(counties_abbrevs)):
    abbrev = counties_abbrevs[i][1][1:]
    states.append(abbrevs_to_states[abbrev])
    abbrevs.append(abbrev)
    counties.append(counties_abbrevs[i][0])

df1["Name"] = pd.Series(counties)
df1["State"] = pd.Series(states)
df1["Abbrev"] = pd.Series(abbrevs)

df1 = df1.reindex(columns=["Name","State","Abbrev","Median Household Income","HS Graduation Rate","College Graduation Rate","Incarceration Rate",
              "Teenage Birth Rate","Population Density","Poverty Rate"]).rename(columns={"Name": "County"})
city_indices = list(df1[df1.County.str.find("city") > -1].index)
for i in range(len(city_indices)):
    city = df1.iloc[city_indices[i],0]
    df1.iat[city_indices[i],0] = city[:-4] + "C" + city[-3:]

df2 = pd.merge(df1, election_df, how="inner", on=["County","State","Abbrev"])
df2 = df2.sort_values(["State","County"])
df2
```

```python
# Life expectancy data: IHME
df3 = pd.read_csv("IHME_USA_LIFE_EXPECTANCY_1985_2010.csv")
df3["Life Expectancy"] = (df3["Female life expectancy (years)"] + df3["Male life expectancy (years)"]) / 2
df3 = df3.reset_index()
df3 = df3.iloc[:,[2,1,11]]
df3["Abbrev"] = df3.State.apply(lambda state: states_to_abbrevs[state])
df4 = pd.merge(df2, df3, how="inner", on=["County", "State", "Abbrev"])
df4
```

```python
# population growth info: US Census
df5 = pd.read_csv("co-est2022-pop.csv")
df5 = df5[df5["April 2020 estimate"].isna() == False]
df5 = df5[df5["Geographic Area"].str[0] == '.']
df5 = df5.iloc[:,[0,2,4]].reset_index()

df5["County Tuple"] = df5["Geographic Area"].str.split(', ')
df5["County"] = df5["County Tuple"].apply(lambda tup: tup[0][1:])
df5["State"] = df5["County Tuple"].apply(lambda tup: tup[1])
df5["Abbrev"] = df5.State.apply(lambda state: states_to_abbrevs[state])

df5["2020"] = df5["2020"].apply(lambda population: int(population.replace(',','')))
df5["2022"] = df5["2022"].apply(lambda population: int(population.replace(',','')))
df5 = df5.iloc[:,[5,6,7,2,3]]
df5 = df5.rename(columns={"2020": "2020 Population", "2022": "2022 Population"})
df5["% Growth"] = (df5["2022 Population"] - df5["2020 Population"]) / df5["2020 Population"]

city_indices = list(df5[df5.County.str.find("city") > -1].index)
for i in range(len(city_indices)):
    city = df5.iloc[city_indices[i],0]
    df5.iat[city_indices[i],0] = city[:-4] + "C" + city[-3:]

df = pd.merge(df4, df5, how="inner", on=["County", "State", "Abbrev"])

originalDF = df.copy()

# normalization
df['Median Household Income'] = (df['Median Household Income'] - df['Median Household Income'].mean()) / df['Median Household Income'].std()
df['Population Density'] = np.log(df['Population Density'])
df = df.rename(columns={"Population Density":"Log Population Density"})
df
```

```python
import matplotlib.pyplot as plt

newDF = df.iloc[:,:-3]
newDF = newDF.dropna()

# Make a dictionary for US geographic region classification and use this to add a new column to the DataFrame
regions = ['Southeast','New England','Midwest','Mid-Atlantic','Southwest','Mountain','Pacific']
len_regions = len(regions) # 7 regions
region_nums = {regions[i]:i for i in range(len_regions)}
colors = ['red','orange','yellow','green','aqua','blue','violet']
region_colors = {regions[i]:colors[i] for i in range(len_regions)}
states_regions = {}
with open("states_regions.txt") as file:
    lines = file.readlines()
    for line in lines:
        arr = line.split(',')
        abbrev, region = arr[0], arr[1].replace('\n','')
        if region not in regions:
            raise Exception(region,'is not a valid region.')
        states_regions[abbrev] = region

newDF['Region'] = newDF['Abbrev'].apply(lambda abbrev: states_regions[abbrev])

# Create a scatterplot of the correlation between each independent variable and life expectancy
independent_vars = list(newDF.columns[3:-2])
dependent_var = 'Life Expectancy'

n_vars = len(independent_vars)
#fig, axs = plt.subplots(n_vars, 1, figsize=(8, 4 * n_vars))
#fig, axs = plt.subplots(2, 2, figsize=(10, 10))  # for a 2x2 grid
fig, axs = plt.subplots(4, 2, figsize=(10, 12))
axs = axs.flatten()
corr_coefs = []

for i, col in enumerate(independent_vars):
    if i >= len(axs):
        break
    
    ax = axs[i]
    curr_plot = ax.scatter(newDF[col], newDF[dependent_var], s=3, c=list(newDF['Region'].map(region_colors)))
    corr = newDF[col].corr(newDF[dependent_var])
    
    # Add a trend line
    m, b = np.polyfit(newDF[col], newDF[dependent_var], 1)
    ax.plot(newDF[col], m*newDF[col] + b, color='black')
    
    # Add title and labels
    ax.set_title(f'{col} vs. {dependent_var}\n(Correlation: {corr:.2f})')
    ax.set_xlabel(col)
    ax.set_ylabel(dependent_var)
    #plt.legend(handles=curr_plot.legend_elements()[0], 
    #           labels=regions,
    #           title="Region")
    corr_coefs.append(corr)

# Adjust layout to prevent overlapping
plt.tight_layout()
# ax.legend(regions) MAKE LEGEND
plt.show()
```

### (c) Can multivariate linear regression provide a somewhat accurate model used to predict life expectancy using the other given variables?

In the following code, we create a multivariate linear regression using all of the factors to use as an initial model. We split the counties into $90\%$ training data and $10\%$ testing data and create a model to fit the training data.

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Define your features and target variable
X = df[independent_vars].copy()
y = df[dependent_var]

# Impute missing values using the mean
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_imputed = imputer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.10, random_state=20)

# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# y_test would be the actual life expectancy values from the test set
# y_pred is the array of predicted life expectancy values from your model

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print('R-squared:', r2)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```

### (d) What is the "best" multivariate linear predictive model for our data? How do we measure "best"?

With $8$ different socio-economic factors, there are $2^{8}-1=255$ different combinations of factors that with which we can form multivariate linear predictive models. In the code below, we create a model for each combination of factors. For each number of factors, we pick the model with the best $r^{2}$ correlation value out of the models with that many factors. This will significantly narrow down our search for the best model.

```python
# Get the coefficients from the model
model_coefs = regr.coef_

# Create a DataFrame to view feature names alongside their corresponding coefficients
feature_names = X.columns
coef_df = pd.DataFrame({'Feature': feature_names, 'Model Coefficient': model_coefs, 'Correlation': corr_coefs})

# Sort the features by the absolute value of their correlation coefficient
sorted_coef_df = coef_df.reindex(coef_df['Correlation'].abs().sort_values().index)

# Print the sorted DataFrame
print(sorted_coef_df)

# Check which combination of factors yields a high r^2 and low MSE while using less factors
n = len(sorted_coef_df)
curr_combo = [True] * n
features_red = []
num_features_red = []
corrs_red = []
mses_red = []

for i in range((2**n)-1):
    curr_included = []
    curr_excluded = []
    for j in range(n):
        if curr_combo[j]:
            curr_included.append(j)
        else:
            curr_excluded.append(j)
    
    least_impactful_features = sorted_coef_df.iloc[curr_excluded,0]
    X_reduced = X.drop(least_impactful_features, axis=1)

    imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
    X_imputed = imputer.fit_transform(X_reduced)

    # Split the data into training and testing sets
    X_train_red, X_test_red, y_train, y_test = train_test_split(X_imputed, y, test_size=0.10, random_state=42)

    # Create a new linear regression object
    regr_red = LinearRegression()

    # Train the model using the training sets with reduced features
    regr_red.fit(X_train_red, y_train)

    # Make predictions using the testing set with reduced features
    y_pred_red = regr_red.predict(X_test_red)

    # Calculate R-squared and Mean Squared Error for the new model
    r2_red = r2_score(y_test, y_pred_red)
    mse_red = mean_squared_error(y_test, y_pred_red)
    
    features_red.append([independent_vars[j] for j in curr_included])
    num_features_red.append(len(curr_included))
    corrs_red.append(r2_red)
    mses_red.append(mse_red)
    
    # updating combination
    feature = n-1
    while feature >= 0 and not curr_combo[feature]:
        curr_combo[feature] = True
        feature -= 1
    curr_combo[feature] = False

# All 2^{8}-1 possible models
red_models_df = pd.DataFrame({'Features': features_red, 'Number of Features': num_features_red,
                              'Correlation': corrs_red, 'Mean Squared Error': mses_red})
sorted_red_models_df = red_models_df.sort_values(by='Correlation', ascending=False)
ax = plt.scatter(sorted_red_models_df['Number of Features'], sorted_red_models_df['Correlation'], s=10)
ax.axes.set_title('Change in Correlation from Adding Features in the Multivariate Linear Model')
ax.axes.set_xlabel('Number of Features')
ax.axes.set_ylabel('Correlation')
plt.show()

# Pick the model with the best correlation for each number of factors
good_models = []
for i in range(n):
    curr_model = sorted_red_models_df[sorted_red_models_df['Number of Features'] == i+1].iloc[0,:]
    good_models.append(curr_model)
good_models_df = pd.DataFrame(good_models) # correlation and MSE start to flatten out after five features

# Get the "best" model
opt_num = 5
best_model = good_models_df.iloc[opt_num-1,:]
print("\nBest Model")
print("    - Features:",best_model['Features'])
print("    - Number of Features:",opt_num)
print("    - Correlation:",best_model['Correlation'])
print("    - Mean Squared Error:",best_model['Mean Squared Error'])
```

### Results

As the plot above shows, the correlation noticeably increases as factors are added, but after 5 factors, the highest correlation in each subset starts to plateau. So, while we desire a larger value for correlation, it is more computationally efficient to reduce the amount of factors when possible. Thus, we choose a model with $5$ factors:
- $x_{1}$: Median Household Income
- $x_{2}$: High School Graduation Rate
- $x_{3}$: Log Population Density
- $x_{4}$: Poverty Rate
- $x_{5}$: Fraction of Democrat Voters in 2020

The equation for this model is: $$\text{Life Expectancy}=0.435330x_{1}-0.316689x_{2}-0.188139x_{3}-6.979230x_{4}+3.403722x_{5}$$

The correlation of this model is $r^{2}=0.68637$, which implies a moderate to strong correlation.

### (e) How can we use this model to predict life expectancy and classify counties based on region?

Using this model, we create a **decision tree** by using the split testing and training data. Then, we compute the precisions, recalls, and F1 scores for each geographic region via a confusion matrix. In particular, the decision tree has a `max_depth` of $10$, which we expect to be high enough to make more accurate in its decisions but not too high as to cause overfitting.

```python
from sklearn.tree import DecisionTreeClassifier

# Define your features and target variable
X = newDF[independent_vars].copy()
y = newDF['Region'].apply(lambda region: region_nums[region])

# Impute missing values using the mean
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_imputed = imputer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.10, random_state=20)

# Create a decision tree and use it to predict the region for the test data
tree = DecisionTreeClassifier(max_depth=10)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)

# Calculate the precision, recall, and F1 score for each region using a confusion matrix
y_test_count = [len(y_test[y_test == i]) for i in range(len_regions)]
y_pred_count = [len(y_pred[y_pred == i]) for i in range(len_regions)]
confusion_matrix = np.zeros([len_regions,len_regions]).astype('int32')

for i in range(len(y_test)):
    prediction = y_pred[i]
    actual = y_test.iloc[i]
    confusion_matrix[actual][prediction] += 1

precisions = np.zeros(len_regions)
recalls = np.zeros(len_regions)
f1_scores = np.zeros(len_regions)

for i in range(len_regions):
    if y_pred_count[i] == 0:
        precisions[i] = 1
    else:
        precisions[i] = confusion_matrix[i][i] / y_pred_count[i]
    if y_test_count[i] == 0:
        recalls[i] = 1
    else:
        recalls[i] = confusion_matrix[i][i] / y_test_count[i]
    if precisions[i] == recalls[i] == 0:
        f1_scores[i] = 0
    else:
        f1_scores[i] = (2 * precisions[i] * recalls[i]) / (precisions[i] + recalls[i])

classification_df = pd.DataFrame({'Region':regions, 'Precision':precisions, 'Recall':recalls, 'F1 Score':f1_scores})
print("Confusion matrix:")
print(confusion_matrix)
classification_df.sort_values(by='F1 Score', ascending=False)
```

### Results

Using the decision tree for our predictions did not yield a very high precision or recall for each region. In particular, the region with the highest for both, the *Southeast*, had a precision of $0.67544$ and a recall of $0.70642$.

### (f) Linear regression did not yield a useful model for logarithmic population density. Will a higher order polynomial regression fit the data better?

As seen in the cell that analyzed linear regression between each socio-economic factor, the logarithmic population density stood out as having very little linear correlation with life expectancy. In this code, we consider fitting it to a quadratic regression instead. We look at both a regular scatter plot of all of the data as well as a binned scatter plot, which only plots the average value of equally spaced bins.

```python
# Create a new dataframe including the life expectancy and log population density to analyze
log_population_df = newDF[['Life Expectancy', 'Log Population Density']]

# Make a binned scatter plot to better visualize data
# Binning the data
log_population_df['x_bin'] = pd.cut(log_population_df['Log Population Density'], bins=100)
grouped = log_population_df.groupby('x_bin')['Life Expectancy'].agg(['mean', 'count']).reset_index()

# Plot the binned scatter
plt.figure(figsize=(10, 6))
plt.scatter(grouped['x_bin'].astype(str), grouped['mean'], s=grouped['count'])
plt.xticks(range(0, len(grouped), 5), grouped['x_bin'].astype(str)[::5], rotation=45)
plt.xlabel('Binned log population density')
plt.ylabel('Average of Life Expectancy in each bin')
plt.title('Binned Scatter Plot of Log Population Density')
plt.ylim(75.5, 80.5)
plt.grid(True)
plt.show()

# Based on the shape shown in the binned scatter plot, do a 2nd degree polynomial fit
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
# Sort the dataframe by log population density for plotting purposes
log_population_df = log_population_df.sort_values(by='Log Population Density')

# Extract the independent variable (log population density) and dependent variable (life expectancy)
x = log_population_df['Log Population Density'].values.reshape(-1, 1)
y = log_population_df['Life Expectancy'].values

# Create polynomial features for the independent variable
poly_features = PolynomialFeatures(degree=2)
x_poly = poly_features.fit_transform(x)

# Fit a linear regression model with polynomial features
model = LinearRegression()
model.fit(x_poly, y)

# Predict values using the model
y_pred = model.predict(x_poly)

# Plot the polynomial regression curve
plt.scatter(x, y, label='Original Data', color='blue', s=5)
plt.plot(x, y_pred, color='red', label='Polynomial Fit (2nd Degree)')
plt.legend()
plt.show()

# Display the regression parameters

# Calculate R-squared value
from sklearn.metrics import r2_score
r_squared = r2_score(y, y_pred)

# Print the regression parameters and R-squared value
equation = f'Life Expectancy = {model.coef_[2]:.2f}*(Log Pop. Density)^2 + {model.coef_[1]:.2f}*(Log Pop. Density) + {model.intercept_:.2f}'
r_squared_text = f'R-squared (R^2) = {r_squared:.4f}'
print(equation)
print(r_squared_text)
```

### Results

While a binned scatter plot indicated that the data may have a somewhat strong correlation to a second degree polynomial, fitting the full data yielded a correlation that was just as bad as in the linear regression case. From this, we can infer that a higher order polynomial regression will likely not fit the data either, so we conclude that **there is no significant correlation between logarithmic population density and life expectancy.**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

socioeconomic_factors = ['Median Household Income', 'HS Graduation Rate', 'College Graduation Rate', 
                         'Incarceration Rate', 'Teenage Birth Rate', 'Log Population Density', 'Poverty Rate']

# Extracting the relevant columns for clustering
X = df[socioeconomic_factors]

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_imputed = imputer.fit_transform(X)

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Determining the optimal number of clusters using the Elbow method
inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow graph to find the optimal number of clusters
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# The elbow in the plot indicates the optimal number of clusters (where inertia starts to decrease more slowly)
# Let's say the elbow is at n_clusters = k, we will use that for the final clustering

# Applying K-means clustering with the optimal number of clusters
optimal_clusters = 4  # replace this with the actual number from the elbow plot
kmeans = KMeans(n_clusters=optimal_clusters, random_state=2)
kmeans.fit(X_scaled)

# Adding the cluster labels to the DataFrame
newDF = originalDF.copy()
newDF['Cluster'] = kmeans.labels_

# Reducing the dimensionality for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plotting the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=newDF['Cluster'], cmap='viridis', marker='o')
plt.title('Clusters based on Socioeconomic Factors')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
